```python
import pandas as pd

# Step 1: Initialize DataFrame
df = pd.DataFrame(columns=['longitude', 'latitude', 'city'])
# Step 2: Read KML File
with open('D:\\FIT3179\\Github\\3179\\test\\Geolocation.kml', 'r') as f:
    city = None
    for line in f:
        # Step 3: Extract City Name
        if '<address>' in line:
            city = line.split('>')[1].split('<')[0].strip()
        # Step 4: Extract Coordinates
        if '<coordinates>' in line:
            coordinates = line.split('>')[1].split('<')[0].strip().split(',')
            longitude = coordinates[0]
            latitude = coordinates[1]
            # Step 5: Append to DataFrame using pd.concat
            new_row = pd.DataFrame({'longitude': [longitude], 'latitude': [latitude], 'city': [city]})
            df = pd.concat([df, new_row], ignore_index=True)

# Step 6: Save to CSV
df.to_csv('D:\\FIT3179\\Github\\3179\\test\\new_geolocation.csv', index=False)

# Display the DataFrame
print(df.head())
```

```python
import pandas as pd

# Load the Excel file
data_file = 'D:\\FIT3179\Github\\3179\\test\\raw_data\\tra-nvs-quarterly-june-2024-2023.xlsx'
xls = pd.ExcelFile(data_file)

# Initialize an empty list to store cleaned data from each sheet
cleaned_data = []

# Iterate through each sheet in the Excel file
for sheet_name in xls.sheet_names:
    # Load the sheet into a DataFrame
    df = pd.read_excel(xls, sheet_name=sheet_name)
    
    # Assuming the actual data starts after a few rows of titles/explanations
    # Here, we try to locate the first row that looks like data
    for i in range(len(df)):
        if df.iloc[i].notna().sum() > 3:  # Assuming a data row has more than 3 non-NA values
            df = df.iloc[i:]
            break

    # Reset column headers if necessary
    df.columns = df.iloc[0]  # Set the first row as header
    df = df[1:]  # Drop the header row from the data
    
    # Drop columns with all NaN values (optional)
    df = df.dropna(axis=1, how='all')
    
    # Drop rows with all NaN values (optional)
    df = df.dropna(axis=0, how='all')
    
    # Ensure the index is unique to avoid reindexing issues
    df.reset_index(drop=True, inplace=True)
    
    # Store cleaned DataFrame in the list
    cleaned_data.append(df)

# Save each cleaned DataFrame to one CSV file
for i, df in enumerate(cleaned_data):
    df.to_csv(f'D:\\FIT3179\\Github\\3179\\test\\raw_data\\{xls.sheet_names[i]}.csv', index=False)
    print(f"Saved {xls.sheet_names[i]}.csv")
    
```

```python
import csv
from geopy.geocoders import Nominatim
import time

# Initialize Nominatim API
geolocator = Nominatim(user_agent="geoapiExercises")

# List of locations
locations = [
    "Sydney", "Melbourne", "Brisbane", "Perth", "Gold Coast", "North Coast NSW", "Hunter", 
    "Sunshine Coast", "South Coast", "Peninsula", "Yarra Valley", "Dandenong Ranges", 
    "Adelaide", "Queensland Country", "Geelong and the Bellarine", "Australia's South West", 
    "Central Coast", "Central NSW", "Blue Mountains", "Gippsland", "Great Ocean Road", 
    "Capital Country", "Tropical North Queensland", "Fleurieu Peninsula", "Bendigo Lodden", 
    "Hobart and the South", "Ballarat", "High Country", "Canberra", "New England North West", 
    "Phillip Island", "Goulburn", "Riverina", "Townsville", "Launceston and the North", 
    "Macedon", "Adelaide Hills", "Capricorn Coast", "North West", "The Murray", 
    "Australia's Coral Coast", "Australia's Golden Outback", "Bundaberg", "Central Murray", 
    "Fraser Coast", "Mackay", "Lakes", "Barossa", "Murray, River, Lakes and Coorong", 
    "Yorke Peninsula", "Limestone Coast"
]

# File path to save the CSV
output_file = 'location_geocodes.csv'

# Open CSV file for writing
with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Location", "Latitude", "Longitude"])  # Writing the headers

    # Fetch latitude and longitude for each location
    for location in locations:
        try:
            location_data = geolocator.geocode(location)
            if location_data:
                writer.writerow([location, location_data.latitude, location_data.longitude])
            else:
                writer.writerow([location, "Not Found", "Not Found"])
        except Exception as e:
            writer.writerow([location, "Error", "Error"])
        time.sleep(1)  # To avoid being blocked by the API

print(f"Geolocation data has been saved to {output_file}")
```
